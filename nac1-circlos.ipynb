```python
#default imports 
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

imagem = cv2.imread('fiapo circlo.png')
imagem = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(imagem,50,150)
circulo = []
 

circulos = cv2.HoughCircles(imagem, cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=200, param2=40, minRadius=80, maxRadius=100)
# circulos = cv2.HoughCircles(imagem, cv2.HOUGH_GRADIENT, dp=1, minDist=20, param1=200, param2=40, minRadius=84, maxRadius=100)

imagem_cinza = cv2.cvtColor(imagem,cv2.COLOR_GRAY2RGB)



if circulos is not None:
    circulos = np.uint16(np.around(circulos))
    for x,y, radius in circulos[0,:]:

        # desenha o contorno do circulo
        cv2.circle(imagem_cinza,(x,y),radius,(255,255,0),5)

        # desenha no centro do circulo
        cv2.circle(imagem_cinza,(x,y),5,(255,0,0),10)
        area = math.pi * math.pow(radius, 2)

    
        circulo.append((x,y))
        print(area)

   


plt.figure(figsize = (10,10))
plt.imshow(imagem_cinza, cmap="Greys_r", vmin=0, vmax=255)
plt.show()


```
